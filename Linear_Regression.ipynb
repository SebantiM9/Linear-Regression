# **Handling Missing Values**

```python
df["horsepower"]=df["horsepower"].replace(to_replace="?",value=np.nan)
df["horsepower"]=df["horsepower"].astype(float)
```

```python
df.info()
```

```python
median1=df["horsepower"].median()
```

```python
df["horsepower"].replace(np.nan,median1,inplace=True)
```
